```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips =sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips)
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day")
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size='size')
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size='size', palette = "YlGnBu")
```

```python
sns.histplot(tips['tip'], bins=15)
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.distplot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x='sex', y='tip', data=tips)
```

[Seaborn Color Palettes and How to Use Them](https://r02b.github.io/seaborn_palettes/)

```python
sns.palplot(sns.color_palette("Blues", 10))
```

```python
sns.barplot(x='sex', y='tip', data=tips, palette = 'Blues')
```

```python
sns.boxplot(x='day', y = 'total_bill', data = tips)
```
