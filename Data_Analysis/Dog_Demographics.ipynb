```python
import matplotlib.pyplot as plt

generation  = ['All', 'Millennials', 'Gen X', "Boomers"]    
percent = [48, 61, 52, 37]

years = ['1946-1964','1965-1980','1981-1996','1997-2012']  # Gen Z  2010-2025

plt.bar(generation, percent, width=0.7, color=('blue', 'red', 'green', 'purple'))
plt.xlabel("Generation Catagory", fontsize = 14)
plt.ylabel("Percent of Dog Ownership", fontsize = 12)
plt.ylim(0, 100)
plt.title("Dog Ownership by Generation Catagory", fontsize = 16)

plt.text(0,percent[0]+2, str(percent[0]) + " %", ha = 'center')
plt.text(1,percent[1]+2, str(percent[1]) + " %", ha = 'center')
plt.text(2,percent[2]+2, str(percent[2]) + " %", ha = 'center')
plt.text(3,percent[3]+2, str(percent[3]) + " %", ha = 'center')

plt.show()
```
